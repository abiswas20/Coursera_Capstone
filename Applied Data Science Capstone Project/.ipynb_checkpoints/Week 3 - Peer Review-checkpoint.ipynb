## PART-1: Scrape table on Wikipedia listing postal codes and boroughs of Toronto neighborhoods

```python
#url to download table from
url="https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M"
```

```python
#using urllib.request to import data
page=urllib.request.urlopen(url)
page
```

```python
soup=BeautifulSoup(page,"lxml")
```

```python
#make source more readable
print(soup.prettify())
```

```python
#table is stored under class 'wikitable sortable'
right_table=soup.find('table',class_='wikitable sortable')
right_table
```

```python
#table values appear between 'td' which themselves are in between 'tr' tags. A,B,C are empty lists to save columns in table.
A=[]
B=[]
C=[]

for row in right_table.findAll('tr'):
    cells=row.findAll('td')
    if len(cells)==3:
        A.append(cells[0].find(text=True))
        B.append(cells[1].find(text=True))
        C.append(cells[2].find(text=True))
```

```python
#Verify what's stored in each list

df_pcode=pd.DataFrame(A)
print(df_pcode)
df_borough=pd.DataFrame(B)
print(df_borough)
df_neighborhood=pd.DataFrame(C)
print(df_neighborhood)
```

```python
#Rename columns and clean-up
d={'Postal_Code':A,'Borough':B,'Neighborhood':C}
df_toronto=pd.DataFrame(data=d)
df_toronto=df_toronto.replace('\\n','',regex=True)

#Select only rows where a Borough is specified
df_toronto=df_toronto[df_toronto['Borough']!='Not assigned']

df_toronto.index.values
#Setting offset so that index starts at 0.
#df_toronto.index-=2 

#unique_postal_codes=df_toronto.index.unique
#unique_postal_codes
#df_toronto.index.values
```

```python
df_toronto.reset_index(drop=True,inplace=True)
df_toronto
```

## PART-2: Add columns showing latitude and longitude of each postal code

```python
#Create empty columns for latitude and longitude
df_toronto['Latitude']=""
df_toronto['Longitude']=""
df_toronto
```

```python
#Reading latitude and longitude from provided link
GeoData=pd.read_csv("https://cocl.us/Geospatial_data")
GeoData
```

```python
#program to match postal code in df_toronto with GeoData and save corresponding latitude and longitude in df_toronto. 
numRows=len(df_toronto)
numRowsG=len(GeoData)
i=0
while i<numRows:
    j=0
    while j<numRowsG:
        if df_toronto.iloc[i,0]==GeoData.iloc[j,0]:
            df_toronto.iloc[i,3]=GeoData.iloc[j,1]
            df_toronto.iloc[i,4]=GeoData.iloc[j,2]
        j+=1
    i+=1
df_toronto    
```

## Create a dataframe with the venues and corresponding categories, latitudes and longitudes

```python
# function that extracts the category of the venue
def get_category_type(row):
    try:
        categories_list = row['categories']  #pull column categories
    except:
        categories_list = row['venue.categories'] #or pull column venue.categories
        
    if len(categories_list) == 0: #if length of list==0, i.e.,there's nothing in categories_list.
        return None
    else:
        return categories_list[0]['name'] #list item #0, dictionary keyword 'name'. extract the actual name from venue.categories.
```

```python
venues = results['response']['groups'][0]['items']  #results['response']['groups'] is a list. 0 is index number.'items' is a dictionary key.

nearby_venues = json_normalize(venues) # flatten JSON
#nearby_venues
# filter columns
filtered_columns = ['venue.name', 'venue.categories', 'venue.location.lat', 'venue.location.lng'] #selecting columns
nearby_venues =nearby_venues.loc[:, filtered_columns] #all rows of selected columns. shows all filtered columns as is.
#print('nearby_venues:\n ',nearby_venues)
#nearby_venues
# filter the category for each row
#print('#1 nearby_venues[venue.categories]:\n',nearby_venues['venue.categories']) #nearby_venues['venue.categories'] shows entire list under venue.categories.
nearby_venues['venue.categories'] = nearby_venues.apply(get_category_type, axis=1) #apply get_category_type across row for each row.
#print('#2 nearby_venues[venue.categories]):\n',nearby_venues['venue.categories'])
# clean columns
nearby_venues.columns = [col.split(".")[-1] for col in nearby_venues.columns] #column name is only the last part of the name with '.' between first and second part.

nearby_venues.head()  #first few rows of the newly cleaned up dataframe
```

## Find top 100 venues for all neighborhoods in toronto_data and group them by neighborhood


```python
def getNearbyVenues(names, latitudes, longitudes, radius=500):
    
    venues_list=[]
    for name, lat, lng in zip(names, latitudes, longitudes):
        print(name)
            
        # create the API request URL
        url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(
            CLIENT_ID, 
            CLIENT_SECRET, 
            VERSION, 
            lat, 
            lng, 
            radius, 
            LIMIT)
            
        # make the GET request
        results = requests.get(url).json()["response"]['groups'][0]['items']
        
        # return only relevant information for each nearby venue
        venues_list.append([(
            name, 
            lat, 
            lng, 
            v['venue']['name'], 
            v['venue']['location']['lat'], 
            v['venue']['location']['lng'],  
            v['venue']['categories'][0]['name']) for v in results])

    nearby_venues = pd.DataFrame([item for venue_list in venues_list for item in venue_list])
    nearby_venues.columns = ['Neighborhood', 
                  'Neighborhood Latitude', 
                  'Neighborhood Longitude', 
                  'Venue', 
                  'Venue Latitude', 
                  'Venue Longitude', 
                  'Venue Category']
    
    return(nearby_venues)
```

```python
# Only name of the neighborhoods in toronto_data are printed.

toronto_venues = getNearbyVenues(names=toronto_data['Neighborhood'],
                                   latitudes=toronto_data['Latitude'],
                                   longitudes=toronto_data['Longitude']
                                  )
toronto_venues
```

## One-hot encoding the presence and type of venues of different venues

```python
# one hot encoding to transform categorical variable to numerical variable.
toronto_onehot = pd.get_dummies(toronto_venues[['Venue Category']], prefix="", prefix_sep="")
toronto_onehot=toronto_onehot[['Neighborhood']+[col for col in toronto_onehot.columns if col!='Neighborhood']]
toronto_onehot['Neighborhood']=toronto_venues['Neighborhood']
toronto_onehot.head()
```

```python
toronto_onehot.shape
```

## one-hot encoding categorical variables, grouped by neighborhood and expressed as normalized frequency

```python
#one-hot encoded categorical variables grouped by neighborhood and expressed as normalized frequency
toronto_grouped = toronto_onehot.groupby('Neighborhood').mean().reset_index()
toronto_grouped
```

```python
toronto_grouped.shape
```

## Top 10 venues in each neighborhood in toronto_grouped dataframe

```python
num_top_venues = 10 #number of venues in each neighborhood to show

for hood in toronto_grouped['Neighborhood']: #go through each neighborhood and list corresponding top 10 venues
    print("----"+hood+"----")
    temp = toronto_grouped[toronto_grouped['Neighborhood'] == hood].T.reset_index() #list of columns written as a column
    temp.columns = ['venue','freq'] 
    temp = temp.iloc[1:]
    temp['freq'] = temp['freq'].astype(float)
    temp = temp.round({'freq': 2})
    print(temp.sort_values('freq', ascending=False).reset_index(drop=True).head(num_top_venues))
    print('\n')
```

## Top 10 categories of venues for each neighborhood in toronto_grouped

```python
def return_most_common_venues(row, num_top_venues):
    row_categories = row.iloc[1:]
    row_categories_sorted = row_categories.sort_values(ascending=False)
    
    return row_categories_sorted.index.values[0:num_top_venues]
```

```python
num_top_venues = 10

indicators = ['st', 'nd', 'rd']

# create columns according to number of top venues
columns = ['Neighborhood']
for ind in np.arange(num_top_venues):
    try:
        columns.append('{}{} Most Common Venue'.format(ind+1, indicators[ind]))
    except:
        columns.append('{}th Most Common Venue'.format(ind+1))

# create a new dataframe
neighborhoods_venues_sorted = pd.DataFrame(columns=columns)
neighborhoods_venues_sorted['Neighborhood'] = toronto_grouped['Neighborhood']
print(toronto_grouped)
for ind in np.arange(toronto_grouped.shape[0]):
    neighborhoods_venues_sorted.iloc[ind, 1:] = return_most_common_venues(toronto_grouped.iloc[ind, :], num_top_venues)

neighborhoods_venues_sorted.head()
```

## Cluster Neighborhoods using K Means Clustering according to Category of top Venues

```python
# set number of clusters
kclusters = 5
#print(toronto_grouped.head())
toronto_grouped_clustering = toronto_grouped.drop('Neighborhood', 1)
print(toronto_grouped_clustering.head())
# run k-means clustering
kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(toronto_grouped_clustering)

# check cluster labels generated for each row in the dataframe
labels=kmeans.labels_[0:10] 
print(labels)
```

```python
# add clustering labels
neighborhoods_venues_sorted.insert(0, 'Cluster Labels', kmeans.labels_)

toronto_merged = toronto_data

# merge toronto_grouped with toronto_data to add latitude/longitude for each neighborhood
toronto_merged = toronto_merged.join(neighborhoods_venues_sorted.set_index('Neighborhood'), on='Neighborhood')

toronto_merged.head()
```

```python
# create map
map_clusters = folium.Map(location=[latitude, longitude], zoom_start=11)

# set color scheme for the clusters
x = np.arange(kclusters)
ys = [i + x + (i*x)**2 for i in range(kclusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, poi, cluster in zip(toronto_merged['Latitude'], toronto_merged['Longitude'], toronto_merged['Neighborhood'], toronto_merged['Cluster Labels']):
    label = folium.Popup(str(poi) + ' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```
