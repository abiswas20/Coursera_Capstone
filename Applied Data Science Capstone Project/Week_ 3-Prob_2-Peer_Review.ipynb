## Read previously saved csv file to re-create df_toronto dataframe

```python
df_toronto=pd.read_csv("toronto_postal_codes.csv",index_col=[0])
df_toronto
```

## Add columns showing latitude and longitude of each postal code

```python
#Create empty columns for latitude and longitude
df_toronto['Latitude']=""
df_toronto['Longitude']=""
df_toronto
```

```python
#Reading latitude and longitude from provided link
GeoData=pd.read_csv("https://cocl.us/Geospatial_data")
GeoData
```

```python
#program to match postal code in df_toronto with GeoData and save corresponding latitude and longitude in df_toronto. 
numRows=len(df_toronto)
numRowsG=len(GeoData)
i=0
while i<numRows:
    j=0
    while j<numRowsG:
        if df_toronto.iloc[i,0]==GeoData.iloc[j,0]:
            df_toronto.iloc[i,3]=GeoData.iloc[j,1]
            df_toronto.iloc[i,4]=GeoData.iloc[j,2]
        j+=1
    i+=1
df_toronto 
```

```python
df_toronto.to_csv("toronto_neighborhoods_latlon.csv", index=False)
```
